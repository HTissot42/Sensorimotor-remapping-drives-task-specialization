```python
def standardize(M) :   
    """ Standardization by column for 2D matrices """
    
    M_copy = np.copy(M)
    M_copy = M_copy.astype(float)
    for j in range(np.shape(M)[1]) :
        if np.std(M[:,j]) != 0 :
            M_copy[:,j] = (M[:,j].astype(float) - float(np.mean(M[:,j])))/float(np.std(M[:,j]))
        else : 
            M_copy[:,j] = M[:,j].astype(float) - float(np.mean(M[:,j]))
            
    return M_copy
```

```python
with open('../Rule-switch/DATA/Dataframes/df_epochs.pickle', 'rb') as f:
    data_epochs = pickle.load(f)

files = ['M03','M04','M05','M07_Region1','M07_Region2','M08_Region1','M08_Region2',\
       'M09_Region1','M09_Region2','M10_Region1','M11_Region1','M11_Region2']

with open('../Response-switch/DATA/Dataframes/df_full.pickle', 'rb') as f:
    data_tc = pickle.load(f)

files_tc = ['M12','M13','M14','M15','M16','M17','M18','M19','M20']
```

```python
act_tensor_tc = []

for f in files_tc :
    data = data_tc[f]
    d1 = data[(data['Context'] == 0)]
    d2 = data[(data['Context'] == 1)]
    act_tensor_tc.append([np.stack(d1['Spike rate']), np.stack(d2['Spike rate'])])

act_tensor_tc = np.array(act_tensor_tc,dtype='object').T
```

```python
act_tensor = []
reg_tensor = []
perf_tensor = []
for t in range(len(data_epochs)):
    act_across_mice = []
    reg_across_mice = []
    perf_across_mice = []
    for f in files :
        data = data_epochs[t][f]
        act_across_mice.append(np.stack(data['Spike rate']))
        
        reg_across_mice.append(standardize(data[['Spatial frequency','Orientation','Choice','Reward']].to_numpy()))

        perf_across_mice.append(np.sum(data['Reward'].to_numpy())/len(data['Reward'].to_numpy()))
        

    act_tensor.append(act_across_mice)
    reg_tensor.append(reg_across_mice)
    perf_tensor.append(perf_across_mice)

act_tensor = np.array(act_tensor,dtype='object')
reg_tensor = np.array(reg_tensor,dtype='object')
perf_tensor = np.array(perf_tensor,dtype='object')
```

```python
colors= [('#666666',0.4),('#5F4830',0.4),('#334E58',0.4)]

def set_color(bplot,colors) :
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)
        #patch.set_alpha(0.4)
        patch.set_edgecolor('black')
        patch.set_linewidth(1.5)

    for median, color in zip(bplot['medians'], colors):
        median.set_color(color)
        median.set_linewidth(4)
        median.set_alpha(1)

    
    for mean, color in zip(bplot['means'], colors):
        mean.set_color(color)
        mean.set_alpha(0.5)
        mean.set_linewidth(1.5)

    for whisk in bplot['whiskers']:
        whisk.set_linewidth(1.5)

    for caps in bplot['caps']:
        caps.set_linewidth(1.5)


delta_rs = np.array([[np.mean(act_tensor[:,f][0]),(np.mean(act_tensor[:,f][4])+np.mean(act_tensor[:,f][7]))/2] for f in range(len(files))])
delta_tc = np.array([(np.mean(act_tensor_tc[:,f][0])+np.mean(act_tensor_tc[:,f][1]))/2 for f in range(len(files_tc))])


values = np.linspace(0.3, 0.8, len(files))

fig,ax = plt.subplots(figsize=(7,7))

for f in range(len(files)):
    plt.scatter(0,delta_rs[f][0],marker='o',c=colors[0],alpha=0.6,s=80)
    plt.scatter(0.5,delta_rs[f][1],marker='o',c=colors[1],alpha=0.6,s=80)

for f in range(len(files_tc)):
    plt.scatter(1,delta_tc[f],marker='o',c=colors[2],alpha=0.6,s=80)

bplot = plt.boxplot([delta_rs[:,0],delta_rs[:,1],delta_tc],positions=[0,0.5,1],showfliers=False,whis=3, patch_artist=True,widths=0.15)

set_color(bplot,colors)

ax.set_ylim(0.002,0.006)
ax.set_xlim(-0.15,1.15)
plt.xticks([0,0.5,1],['Passive','Rule-switch','Response-switch'],fontsize=20,rotation=8)
plt.yticks([0.002,0.004,0.006],fontsize=20)
plt.ylabel('Inferred spike rate (A.U)',fontsize=22)
ax.spines[['right', 'top']].set_visible(False)

"""
leg = plt.legend(ncol=2,frameon=False,fontsize=11)
for line in leg.get_lines():
    line.set_linewidth(3)
"""

plt.savefig('Plots/SVG/cross_epochs.SVG', dpi = 300,bbox_inches='tight')
plt.savefig('Plots/PNG/cross_epochs.PNG', dpi = 300,bbox_inches='tight')
```
