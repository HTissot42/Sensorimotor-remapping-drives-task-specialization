```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import pickle
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
import matplotlib as mpl

mpl_config = pd.read_csv('../../mpl_config.csv').to_dict(orient='records')[0]

mpl.rcParams.update(mpl_config)
```

```python
def standardize(M) :   
    """ Standardization by column for 2D matrices """
    
    M_copy = np.copy(M)
    M_copy = M_copy.astype(float)
    for j in range(np.shape(M)[1]) :
        if np.std(M[:,j]) != 0 :
            M_copy[:,j] = (M[:,j].astype(float) - float(np.mean(M[:,j])))/float(np.std(M[:,j]))
        else : 
            M_copy[:,j] = M[:,j].astype(float) - float(np.mean(M[:,j]))
            
    return M_copy


def full_mean(d, target = 'Spike rate') :
    """Compute mean firing rates for each experimental condition (10)"""
    labels = [k for k in range(1,11)]
    
    mean_rates = []
    
    for label in labels :
        d_label = d[d['Label'] == label]
        
        mean_rates.append(np.mean(np.stack(d_label[target]),axis=0))
    
    return mean_rates

def sample_spherical(ndim=10):
    """Draw a random multivariate gaussian of dimension ndim"""
    vec = np.random.randn(ndim)
    vec = unit_vector(vec)
    return vec

def unit_vector(vector):
    """ Returns the unit vector of the vector.  """
    return vector / np.linalg.norm(vector)
```

```python
files = ['M12','M13','M14','M15','M16','M17','M18','M19','M20']

with open('../DATA/Dataframes/df_full.pickle', 'rb') as f:
    data_full = pickle.load(f)
```

### PCA model trained on mean over post stimulus time window for each condition (assessing inter-condition variance)

```python
mean_rates = [full_mean(data_full[file],target='Spike rate') for file in files]
M_mean = np.concatenate(mean_rates, axis = 1)
M_mean = standardize(M_mean)

tot_var = np.sum(np.std(M_mean,axis=0))
print(tot_var)

n_random_vec = 10000
random_vecs = []
for n in range(n_random_vec) :
    random_vecs.append(sample_spherical(np.shape(M_mean)[1]))

random_vecs = np.array(random_vecs).T
random_expvar = np.std(M_mean@random_vecs,axis=0)**2/tot_var
```

```python
plt.figure(figsize=(12,7))


pca = PCA(n_components = 10)
M_mean_pca = pca.fit_transform(M_mean)

plt.plot(range(1,11),np.std(M_mean_pca,axis=0)**2/tot_var,color='midnightblue',linewidth=3)

plt.ylabel('% Explained variance')

plt.xlabel('# PCs')
plt.xticks(range(1,11))
plt.yticks([0,0.1,0.2,0.3],['0','10%','20%','30%'])
plt.xlim(1,10)
plt.ylim(0,0.31)

plt.fill_between(range(1,11),y1 = 0, y2 = 2*np.percentile(random_expvar,99.999), color='black', alpha = 0.1) #For visualisation purpose

plt.axvline(9,linestyle='--',color='black')

"""
plt.savefig('Plots/SVG/PCA_cond.svg', dpi = 300,bbox_inches='tight')
plt.savefig('Plots/PNG/PCA_cond.PNG', dpi = 300,bbox_inches='tight')
"""
```

### cvPCA

```python
nb_cv = 20 #Number of train-test splits 
nb_comp = 30

train_size = int(len(M_mean)*0.5)
data_idx = [k for k in range(len(M_mean))]


test_std = []
train_std = []

random_train_std = []
for n in tqdm(range(nb_cv)):

    train_idx = np.sort(np.random.choice(data_idx, train_size, replace = False))
    test_idx = np.array([idx for idx in data_idx if idx not in train_idx])

    M_train = M_mean[train_idx,:]
    M_train = standardize(M_train)
    
    M_test = M_mean[test_idx,:]
    M_test = standardize(M_test)
    
    cvpca = PCA(n_components = nb_comp)
    M_train_pca = cvpca.fit_transform(M_train)

    M_test_pca = cvpca.transform(M_test)

    # Computing variance explained
    train_tot_var = np.sum(np.std(M_train,axis=0)**2)
    test_tot_var = np.sum(np.std(M_test,axis=0)**2)
    
    train_std.append(np.std(M_train_pca,axis=0)**2/train_tot_var)
    test_std.append(np.std(M_test_pca,axis=0)**2/test_tot_var)

    # Comparing data explained to random vectors
    n_random_vec = 10000
    random_vecs = []
    for n in range(n_random_vec) :
        random_vecs.append(sample_spherical(np.shape(M_test)[1]))
    
    random_vecs = np.array(random_vecs).T
    random_expvar = np.std(M_test@random_vecs,axis=0)**2/test_tot_var
    random_train_std.append(random_expvar)
    

test_std = np.array(test_std)
train_std = np.array(train_std)

random_train_std = np.array(random_train_std)

random_mean = np.mean(random_train_std)
random_std = np.std(random_train_std)

expvar_5th = np.percentile(random_train_std,0)
expvar_95th = np.percentile(random_train_std,99.999)
```

```python
plt.figure(figsize = (12,7))

plt.plot(range(1,nb_comp+1),np.mean(train_std,axis=0), label = 'Train sets',color='#58B09C',linewidth=3)
plt.plot(range(1,nb_comp+1),np.mean(test_std,axis=0), label ='Test sets',color='#386150',linewidth=3)

plt.fill_between(range(1,nb_comp+1), np.mean(train_std,axis=0) - np.std(train_std,axis=0)/np.sqrt(nb_cv),\
                np.mean(train_std,axis=0) + np.std(train_std,axis=0)/np.sqrt(nb_cv), color = '#58B09C',alpha=0.15)

plt.fill_between(range(1,nb_comp+1), np.mean(test_std,axis=0) - np.std(test_std,axis=0)/np.sqrt(nb_cv),\
                np.mean(test_std,axis=0) + np.std(test_std,axis=0)/np.sqrt(nb_cv), color = '#386150',alpha=0.15)

plt.xlim(1,nb_comp + 1)
plt.ylim((0,0.1))
plt.xlabel('# PCs')
plt.ylabel('% Explained variance')
plt.legend(frameon=False)

#plt.yticks([0,0.05,0.1,0.15],['0','5%','10%','15%'])
plt.yticks([0,0.05,0.1],['0','5%','10%'])
plt.xticks([1,5,10,15,20,25,30])

plt.fill_between(range(nb_comp+1),y1 = 0, y2 = 2*expvar_95th, color='black', alpha = 0.1)

plt.axvline(5,linestyle='--',color='black',linewidth=1.5)

"""
plt.savefig('Plots/SVG/cvPCA_time.svg', dpi = 300,bbox_inches='tight')
plt.savefig('Plots/PNG/cvPCA_time.PNG', dpi = 300,bbox_inches='tight')
"""
```

#### With Covariance

```python
## Assessing number of components to keep using covariance of projections over PCs (inspired by Bimbard et al 2022)
## This method did not yield a smooth decrease of covariances across components which made difficult the choice of a proper boundary.
## Cross-validation on variance explained (see above) was then favored.

nb_cv = 20
nb_comp = 30

train_size = int(len(M_mean)*0.5)
data_idx = [k for k in range(len(M_mean))]


train_test_cov = []
train_test_cov_random = []

random_train_std = []
for n in tqdm(range(nb_cv)):
    train_idx = np.sort(np.random.choice(data_idx, train_size, replace = False))
    test_idx = np.array([idx for idx in data_idx if idx not in train_idx])

    M_train = M_mean[train_idx,:]
    M_train = standardize(M_train)
    
    M_test = M_mean[test_idx,:]
    M_test = standardize(M_test)
    
    cvpca = PCA(n_components = nb_comp)
    M_train_pca = cvpca.fit_transform(M_train)

    M_test_pca = cvpca.transform(M_test)

    # Computing covariances
    tt_cov = np.cov(M_train_pca.T, M_test_pca.T)
    train_test_cov.append([tt_cov[i,nb_comp+i] for i in range(nb_comp)])

    n_random_vec = 10000
    random_vecs = []
    for n in range(n_random_vec) :
        random_vecs.append(sample_spherical(np.shape(M_test)[1]))
    
    random_vecs = np.array(random_vecs).T
    random_tt_cov = np.cov((M_train@random_vecs).T, (M_test@random_vecs).T)
    train_test_cov_random.append([random_tt_cov[i,len(random_vecs)+i] for i in range(len(random_vecs))])
    

train_test_cov = np.array(train_test_cov)
train_test_cov_random = np.array(train_test_cov_random)


expvar_0th = np.percentile(train_test_cov_random,0)
expvar_99th = np.percentile(train_test_cov_random,99.999)
```
