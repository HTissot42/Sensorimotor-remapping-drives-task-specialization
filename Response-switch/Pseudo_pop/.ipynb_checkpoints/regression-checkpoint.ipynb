```python
def standardize(M) :   
    """ Standardization by column for 2D matrices """
    
    M_copy = np.copy(M)
    M_copy = M_copy.astype(float)
    for j in range(np.shape(M)[1]) :
        if np.std(M[:,j]) != 0 :
            M_copy[:,j] = (M[:,j].astype(float) - float(np.mean(M[:,j])))/float(np.std(M[:,j]))
        else : 
            M_copy[:,j] = M[:,j].astype(float) - float(np.mean(M[:,j]))
            
    return M_copy

def unit_vector(vector):
    """ Returns the unit vector of the vector.  """
    return vector / np.linalg.norm(vector)


def angle_between(v1, v2, directed = False):
    """ Returns the angle in radians between vectors 'v1' and 'v2'::

            >>> angle_between((1, 0, 0), (0, 1, 0))
            1.5707963267948966
            >>> angle_between((1, 0, 0), (1, 0, 0))
            0.0
            >>> angle_between((1, 0, 0), (-1, 0, 0))
            3.141592653589793
    """
    v1_u = unit_vector(v1)
    v2_u = unit_vector(v2)
    delta = np.arccos(np.dot(v1_u, v2_u))*180/np.pi

    if not directed : 
        if delta > 90 :
            delta = 180 - delta
    return delta

def sample_spherical(ndim=10):
    """Draw random unit vectors of ndim dimensions from multivariate gaussian"""
    vec = np.random.randn(ndim)
    vec = unit_vector(vec)
    return vec

def smooth(x,y,nb_point) :
    """Smooth a (x,y) serie by inserting nb_point with spline interpolation"""
    x_smooth = np.linspace(np.min(x), np.max(x), nb_point) 
    y_smooth = make_interp_spline(x, y, k=3)(x_smooth)

    return x_smooth, y_smooth

def white_grid(axs, dim) :
    """Function to render matrices"""
    axs.spines[:].set_visible(False)
    axs.set_xticks(np.arange(dim)-.5, minor=True)
    axs.set_yticks(np.arange(dim)-.5, minor=True)
    axs.grid(which="minor", color="w", linestyle='-', linewidth=3)
    axs.tick_params(which="minor", bottom=False, left=False)
    
def bootstrap(x, n_bootstrap=1000) :
    """Bootstrap method to obtain confidence interval on the mean of x"""
    sample_size = len(x)
    bootstraps = []
    for k in range(n_bootstrap) :
        bootstraps.append(np.mean(np.random.choice(x,sample_size,replace=True)))

    return (np.percentile(bootstraps,2.5),np.percentile(bootstraps,97.5))
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import KFold, GridSearchCV, RepeatedKFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# Parameter grid for regularization
alphas = [(10**(-5))*np.exp(k/5) for k in range(1,100)]

def compute_R2_stats(alpha,X,Y,permutation,nb_fold) :
    """Compute corrected variance explained (cvR2/deltaR2) for each regressors (permutation test)"""
    pipe = Pipeline([('scaler',  StandardScaler()),('Ridge', Ridge(alpha=alpha,fit_intercept=True))])
        
    kf_permu = KFold(n_splits = nb_fold, shuffle = True, random_state = 42)
    
    output = cross_validate(pipe,X,Y,cv=kf_permu,error_score=0, return_estimator=True)
    scores = output['test_score']
    m = output['estimator'][np.argmax(scores)]
    y_pred = m.predict(X)
    MSE = np.mean((Y-y_pred)**2, axis = 0)
    R2 = 1-MSE    #Y variance is set to 1 after Z-scoring
    
    X_shuffle = X[permutation,:]
    
    output = cross_validate(pipe,X_shuffle,Y,cv=kf_permu,error_score=0, return_estimator=True)
    scores = output['test_score']
    m_shuffle = output['estimator'][np.argmax(scores)]
    y_pred_shuffle = m_shuffle.predict(X_shuffle)
    MSE_shuffle = np.mean((Y-y_pred_shuffle)**2, axis = 0)
    R2_shuffle = 1-MSE_shuffle
    
    cv_R2 = []
    delta_R2 = []
    
    for i in range(np.shape(X)[1]) :
        X_cv = np.copy(X)
        X_cv[:,:i] = X_cv[permutation,:i]
        X_cv[:,i+1:] = X_cv[permutation,i+1:]

        output = cross_validate(pipe,X_cv,Y,cv=kf_permu,error_score=0, return_estimator=True)
        scores = output['test_score']
        
        m_cv = output['estimator'][np.argmax(scores)]
        y_pred_cv = m_cv.predict(X_cv)   
        MSE_cv = np.mean((Y-y_pred_cv)**2, axis = 0)
        R2_cv = 1-MSE_cv
        cv_R2.append(R2_cv - R2_shuffle)
        
        X_delta = np.copy(X)
        X_delta[:,i] = X_delta[permutation,i]
        
        output = cross_validate(pipe,X_delta,Y,cv=kf_permu,error_score=0,return_estimator = True)
        scores = output['test_score']
        
        m_delta = output['estimator'][np.argmax(scores)]
        y_pred_delta = m_delta.predict(X_delta)
        MSE_delta = np.mean((Y-y_pred_delta)**2, axis = 0)
        R2_delta = 1-MSE_delta
        delta_R2.append(R2 - R2_delta)
        
    return np.array(cv_R2),np.array(delta_R2)


def get_best_alpha(X,Y,nb_fold=5, return_grid = False) : 
    """Method that fit hyperparameter alpha for Ridge regression"""
    pipe = Pipeline([('scaler',  StandardScaler()),('Ridge', Ridge(fit_intercept=True))])    
    
    kf = RepeatedKFold(n_splits = nb_fold,n_repeats = 10)
    

    parameters = {'Ridge__alpha' : alphas}
    
    clf = GridSearchCV(pipe, param_grid = parameters, scoring = 'r2', cv = kf)

    clf.fit(X,Y)
    
    best_alpha = clf.best_params_['Ridge__alpha']
    
    if return_grid :
        return clf, best_alpha
    else :
        return best_alpha
    
    
def compute_regression(X, Y, alpha = None, nb_fold = 5,nb_repeat=10, nb_permu = 0, stats = False, return_permu_models = False) :
    """Method that perform ridge regression of Y according to X.
    Additionnaly if nb_permu is more than 0 and stats si True 
    the method will compute chance corrected (cvR2/deltaR2) variance explained for each regressor
    """
    if alpha == None :
        best_alpha = get_best_alpha(X,Y)
    
    else :
        best_alpha = alpha
    

    pipe = Pipeline([('scaler',  StandardScaler()),('Ridge', Ridge(alpha=best_alpha,fit_intercept=True))])
    
    rkf = RepeatedKFold(n_splits = nb_fold, n_repeats = nb_repeat)
    output = cross_validate(pipe,X,Y,cv=rkf,error_score=0, return_estimator=True)
    models = output['estimator']
    scores = output['test_score']
        
    if nb_permu > 0 :
        N = np.shape(X)[0]
        
        delta_R2s = []
        cv_R2s = []
        permu_models = []
        for p in tqdm(range(nb_permu)) :
            
            permutation = np.random.permutation(N)
            
            if stats :
                cv_R2, delta_R2 = compute_R2_stats(best_alpha,X,Y,permutation,nb_fold)
                delta_R2s.append(delta_R2)
                cv_R2s.append(cv_R2)
            
            if return_permu_models :
                permutation = np.random.permutation(N)
                X_shuffle = X[permutation,:]
        
                pipe = Pipeline([('scaler',  StandardScaler()),('Ridge', Ridge(alpha=best_alpha,fit_intercept=True))])
                    
                kf_permu = KFold(n_splits = nb_fold, shuffle=True)
                output = cross_validate(pipe,X_shuffle,Y,cv=kf_permu,error_score=0, return_estimator=True)
                scores = output['test_score']
                ms_permu = output['estimator']
                permu_models.append([ms_permu[k]['Ridge'] for k in range(nb_fold)])
        
        
        if stats&return_permu_models :
            return models,scores, permu_models, np.array(delta_R2s), np.array(cv_R2s)
        
        if stats : 
            return models,scores, np.array(delta_R2s), np.array(cv_R2s)
        
        if return_permu_models :
            return models,scores, permu_models
        

    return models,scores
```

```python
## Akaike information criterion to determine if task-dependent or task-independent model is best to describe data
plt.figure(figsize=(10,7))

def AIC(N,K,r2) :
    return 2*K + N*(np.log(2*np.pi*(1-r2)) + 1)


K1 = np.shape(X)[1]
K2 = np.shape(simple_regressor)[1]
N = np.shape(X)[0]

test_scores1 = np.array([grid1.cv_results_['split' + str(k) + '_test_score'] for k in range(50)])
test_scores2 = np.array([grid2.cv_results_['split' + str(k) + '_test_score'] for k in range(50)])

print(np.shape(test_scores1))

AICs1 = AIC(N,K1,test_scores1)
AICs2 = AIC(N,K1,test_scores2)

plt.plot(alphas, np.mean(AICs2, axis = 0),label='Task-independent', color = 'orangered', alpha = 0.7)
plt.plot(alphas, np.mean(AICs1, axis = 0),label='Task-dependent', color = 'darkblue', alpha = 0.7)


plt.fill_between(alphas, np.mean(AICs1, axis = 0) - np.std(AICs1, axis = 0)*1.96/np.sqrt(len(AICs1)), np.mean(AICs1, axis = 0) + np.std(AICs1, axis = 0)*1.96/np.sqrt(len(AICs1)), color = 'blue', alpha = 0.15)
plt.fill_between(alphas, np.mean(AICs2, axis = 0) - np.std(AICs2, axis = 0)*1.96/np.sqrt(len(AICs2)), np.mean(AICs2, axis = 0) + np.std(AICs2, axis = 0)*1.96/np.sqrt(len(AICs2)), color = 'orangered', alpha = 0.15)


plt.xscale('log')
plt.xlabel(r'$\alpha$')
plt.ylabel('AIC')
plt.legend(frameon = False)

plt.savefig('Plots/PNG/AIC.PNG', dpi = 300,bbox_inches='tight')
plt.savefig('Plots/SVG/AIC.SVG', dpi = 300,bbox_inches='tight')
```

### Angles over split of pseudo-population

Original pipeline to compute angles between regression axes. This method was abandoned for a similar analysis leveraging individual data instead of pseudo-population data for better interpretability. Both methods yielded consistent results

```python
d = merged_data

M = np.stack(d['Spike rate'])
M = standardize(M)
M = pca1.transform(M)

simple_regressor = d[['Category ID', 'Choice','Motor action','Context','Reward']].to_numpy()
simple_regressor = standardize(simple_regressor)
interactions = pd.concat((simple_regressor[:,0]*(d['Context'] == 0), simple_regressor[:,1]*(d['Context'] == 0),\
                          simple_regressor[:,0]*(d['Context'] == 1), simple_regressor[:,1]*(d['Context'] == 1), simple_regressor[:,2]*(d['Context'] == 1)),axis=1)

interactions = interactions.to_numpy()


X = np.concatenate((interactions,simple_regressor[:,3:]),axis=1)
Y = M

plot_labels = ['Category G/NG','Choice G/NG','Category L/R','Choice L/R','Lick L/R','Task']

#grid, best_alpha = get_best_alpha(X,Y, nb_fold = 10, return_grid = True)
```

```python
trials_per_cond = d.groupby(['Label']).count()['Category ID'].to_numpy()

n_splits = 5 # Number of splits used to train different regression models

idxs_cond = [np.array(range(trials_per_cond[c])) for c in range(len(trials_per_cond))]
for idxs in idxs_cond :
    np.random.shuffle(idxs)

X_splits = []
Y_splits = []
for i in range(n_splits) :
    X_split = []
    Y_split = []
    for c in range(len(trials_per_cond)) :
        idxs = idxs_cond[c]
        
        X_cond = X[np.sum(trials_per_cond[:c]):np.sum(trials_per_cond[:c+1])]

        n_rows = len(X_cond)
        rows_per_split = int(n_rows/n_splits)
        
        X_cond_split = np.array([X_cond[idx] for idx in idxs[rows_per_split*i:rows_per_split*(i+1)]])

        Y_cond = Y[np.sum(trials_per_cond[:c]):np.sum(trials_per_cond[:c+1])]
        Y_cond_split = np.array([Y_cond[idx] for idx in idxs[rows_per_split*i:rows_per_split*(i+1)]])
        
        X_split.append(X_cond_split)
        Y_split.append(Y_cond_split)

    print(X_split[0].shape)
    print(X_split[1].shape)
    X_splits.append(np.concatenate(X_split,axis=0))
    Y_splits.append(np.concatenate(Y_split,axis=0))
    

pipe = Pipeline([('scaler',  StandardScaler()),('Ridge', Ridge(alpha=best_alpha,fit_intercept=True))])


reg_angle = []
for i in range(n_splits) :
    pipe.fit(X_splits[i],Y_splits[i])

    reg_angle.append(pipe['Ridge'].coef_.T)
```

## Regression across time

```python
## We perform here multiple ridge regression on different time periods from trials. 
## This makes us able to 1) inspect variations of variance explained through a given trial
## 2) Study dynamic change of regression axes geometry (i.e. population representations)

d = merged_data

D = np.stack(d['Trajectory'])  ## This time we extract time data instead of averages.

regressor_labels = ['Category','Choice','Motor action','Reward','Context','Category G/NG','Choice G/NG', 'Category L/R','Choice L/R','Motor action L/R']

simple_regressor = d[['Category ID', 'Choice','Motor action','Context','Reward']].to_numpy()
simple_regressor = standardize(simple_regressor)
interactions = pd.concat((simple_regressor[:,0]*(d['Context'] == 0), simple_regressor[:,1]*(d['Context'] == 0),\
                          simple_regressor[:,0]*(d['Context'] == 1), simple_regressor[:,1]*(d['Context'] == 1), simple_regressor[:,2]*(d['Context'] == 1)),axis=1)

interactions = interactions.to_numpy()


X = np.concatenate((interactions,simple_regressor[:,3:]),axis=1)

X = standardize(X)

reg_t = []
scores_t = []
deltaR2_t = []
cvR2_t = []
for t in tqdm(range(np.shape(D)[1])) :
    M = D[:,t,:]
    M = standardize(M)
    M = pca2.transform(M)
    

    inverse_transform = True
    if inverse_transform :
        Y = standardize(pca2.inverse_transform(M))
    else :
        Y = standardize(M)
    
    reg, scores, deltaR2, cvR2  = compute_regression(X,Y, alpha=1, nb_fold = 5, nb_permu = 100, stats = True, return_permu_models=False)

    reg_t.append(reg)
    scores_t.append(scores)
    deltaR2_t.append(deltaR2)
    cvR2_t.append(cvR2)

reg_t = np.array(reg_t)
scores_t = np.array(scores_t)
deltaR2_t = np.array(deltaR2_t)
cvR2_t = np.array(cvR2_t)
```

```python
def compute_corrmat(s1_t,s2_t,mean=True) :
    """Compute cross-correlation matrices between two time series"""
    corr_mat = np.zeros(shape=(len(s1_t),len(s2_t)))
    for t in range(len(s1_t)) :
        for k in range(len(s2_t)) :
            if mean :
                corr_mat[t,k] = np.mean(np.array([pearsonr(s1_t[t,j],s2_t[k,j])[0] for j in range(np.shape(s1_t)[1])]))
            else :
                corr_mat[t,k] = pearsonr(s1_t[t],s2_t[k])[0]

    return corr_mat

def compute_covmat(s_t, mean=True) :
    """Compute covariance matrices between two time series"""
    if mean :
        cov_mat = np.mean(np.array([np.cov(s_t[:,j,:]) for j in range(np.shape(s_t)[1])]),axis=0)
    else :
        cov_mat = np.cov(s_t)

    return cov_mat
```
