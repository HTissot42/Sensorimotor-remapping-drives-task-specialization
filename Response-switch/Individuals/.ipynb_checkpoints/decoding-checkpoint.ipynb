```python
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
import matplotlib as mpl
import pyvista as pv
from scipy.stats import mannwhitneyu, wilcoxon

mpl_config = pd.read_csv('../../mpl_config.csv').to_dict(orient='records')[0]

mpl.rcParams.update(mpl_config)
```

```python
def standardize(M) :   
    """ Standardization by column for 2D matrices """
    
    M_copy = np.copy(M)
    M_copy = M_copy.astype(float)
    for j in range(np.shape(M)[1]) :
        if np.std(M[:,j]) != 0 :
            M_copy[:,j] = (M[:,j].astype(float) - float(np.mean(M[:,j])))/float(np.std(M[:,j]))
        else : 
            M_copy[:,j] = M[:,j].astype(float) - float(np.mean(M[:,j]))
            
    return M_copy


def produce_trial_index(d) :
    """Produce a dictionnary of trial indices corresponding to each label."""
    
    trial_idx = None
    
    go_hit_idx = (d['Choice'] == 1)&(d['Category ID'] == 2)&(d['Context'] == 0)
    go_error_idx = (d['Choice'] == 1)&(d['Category ID'] == 1)&(d['Context'] == 0)

    nogo_hit_idx = (d['Choice'] == -1)&(d['Category ID'] == 1)&(d['Context'] == 0)
    nogo_error_idx = (d['Choice'] == -1)&(d['Category ID'] == 2)&(d['Context'] == 0)

    right_hit_idx = (d['Choice'] == 1) & (d['Category ID'] == 2)&(d['Context'] == 1)
    left_hit_idx = (d['Choice'] == -1) & (d['Category ID'] == 1)&(d['Context'] == 1)

    right_error_idx = (d['Choice'] == 1) & (d['Category ID'] == 1)&(d['Context'] == 1)
    left_error_idx = (d['Choice'] == -1) & (d['Category ID'] == 2)&(d['Context'] == 1)

    # Snooze : The animal did not respond in the 2AFC paradigm.
    snooze_right_idx = (d['Choice'] == 0)&(d['Category ID'] == 2)&(d['Context'] == 1)
    snooze_left_idx = (d['Choice'] == 0)&(d['Category ID'] == 1)&(d['Context'] == 1)

    trial_idx = {'Go / Go : Hit' : go_hit_idx, 'Go / Nogo : Miss':nogo_error_idx, 'NoGo / NoGo : CR' : nogo_hit_idx , \
              'NoGo / Go : FA' : go_error_idx,  'Left / Left':left_hit_idx, 'Left / Right':left_error_idx, 'Left / Snooze':snooze_left_idx, \
           'Right / Right':right_hit_idx, 'Right / Left':right_error_idx, 'Right / Snooze':snooze_right_idx}

    return trial_idx


def set_color(bplot,colors) :
    """Method to render box plots"""
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)
        #patch.set_alpha(0.4)
        patch.set_edgecolor('black')
        patch.set_linewidth(1.4)

    for median, color in zip(bplot['medians'], colors):
        median.set_color(color)
        median.set_linewidth(4)
        median.set_alpha(1)

    
    for mean, color in zip(bplot['means'], colors):
        mean.set_color(color)
        mean.set_alpha(0.5)
        mean.set_linewidth(1.5)

    for whisk in bplot['whiskers']:
        whisk.set_linewidth(1.4)

    for caps in bplot['caps']:
        caps.set_linewidth(1.4)
```

```python
with open('../DATA/Dataframes/df_pseudo.pickle', 'rb') as f:
    merged_data = pickle.load(f)

with open('../Pseudo_pop/Models/pca_notime.pickle', 'rb') as f:
    pca1 = pickle.load(f)

files = ['M12','M13','M14','M15','M16','M17','M18','M19','M20']

with open('../DATA/Dataframes/df_full.pickle', 'rb') as f:
    data_full = pickle.load(f)
```

```python
neuron_per_animal = []
for file in files :
    neuron_per_animal.append(len(data_full[file]['Spike rate'][0]))

def split_PCA(M) :
    # Return PCA transform matrix for each animal (Projection of pseudo-population PCA matrix into the subspace of each animal)
    M_splitted = []
    c=0
    for i in range(len(neuron_per_animal)) :
        M_splitted.append(M[:,c:c+neuron_per_animal[i]])
        c+=neuron_per_animal[i]

    return M_splitted


pca_s = split_PCA(pca1.components_)
```

```python
SF_s = np.array([0.023,0.027,0.033,0.06,0.1,0.25])
ORI_s = np.array([0,15,30,60,75,90])

for file in files :
    d = data_full[file]
    relevant = ((d['Stim ID']-1)//6).to_numpy()
    irrelevant = ((d['Stim ID']-1)%6).to_numpy()

    d['Stim ID relevant'] = relevant 
    d['Stim ID irrelevant'] = irrelevant 
    
    if d['Active rule'][0] == 'SF' :
        d['SF'] = SF_s[relevant]
        d['ORI'] = ORI_s[irrelevant]

    elif d['Active rule'][0] == 'ORI' :
        d['SF'] = SF_s[irrelevant]
        d['ORI'] = ORI_s[relevant]
```

## Stim decoding

```python
from sklearn.model_selection import KFold, GridSearchCV, RepeatedKFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC



def decoding_intra_cat(df,pca_m) :
    cat1_slice = (df['Stim ID relevant'] == 0) + (df['Stim ID relevant'] == 2)
    cat2_slice = (df['Stim ID relevant'] == 3) + (df['Stim ID relevant'] == 5)

    Y1 = df['Stim ID relevant'][cat1_slice]
    Y2 = df['Stim ID relevant'][cat2_slice]
    
    M1 = np.stack(df['Spike rate'][cat1_slice])
    X1 = standardize(M1)
    #X1 = standardize(M1@pca_m)
    
    M2 = np.stack(df['Spike rate'][cat2_slice])
    X2 = standardize(M2)
    #X2 = standardize(M2@pca_m)

    pipe = Pipeline([('scaler',  StandardScaler()),('svc', SVC())])
    rkf = RepeatedKFold(n_splits = 2, n_repeats = 10)
    fit1 = cross_validate(pipe,X1,Y1,cv=rkf,error_score=0, return_estimator=True)
    fit2 = cross_validate(pipe,X2,Y2,cv=rkf,error_score=0, return_estimator=True)

    return [fit1['test_score'],fit2['test_score']]


scores_intra_cat = []
for f in range(len(files)) :
    print(f)
    d = data_full[files[f]]
    pca_mat = pca_s[f].T

    scores_intra_cat.append(decoding_intra_cat(d,pca_mat))

scores_intra_cat = np.array(scores_intra_cat)
```

```python
plt.figure(figsize=(10,8))

plt.axhline(0.5,linestyle='--',color='black',alpha=0.5)

plt.fill_between(range(-1,len(files)+1),0.5,alpha=0.05,color='black')

bplot1=plt.boxplot(scores_intra_cat[:,0,:].T,positions=[-0.2+k for k in range(len(files))],patch_artist=True,bootstrap=1000,sym='',widths=.3)
bplot2=plt.boxplot(scores_intra_cat[:,1,:].T,positions=[0.2+k for k in range(len(files))],patch_artist=True,bootstrap=1000,sym='',widths=.3)

set_color(bplot1,['blue']*len(files))
set_color(bplot2,['red']*len(files))


plt.xlim(-0.5,len(files)-0.5)
plt.xticks(range(len(files)),files)


plt.ylim(0.3,1)
plt.ylabel('Decoding accuracy')
plt.yticks([0.5,0.6,0.7,0.8,0.9,1],['50%','60%','70%','80%','90%','100%'])


plt.savefig('Plots/SVG/stim_decod.SVG', dpi = 300,bbox_inches='tight')
plt.savefig('Plots/PNG/stim_decod.PNG', dpi = 300,bbox_inches='tight')
```

```python
from sklearn.model_selection import KFold, GridSearchCV, RepeatedKFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC



def decoding_cat(df,pca_m) :
    GNG_slice = (df['Context'] == 0)
    LR_slice = (df['Context'] == 1)

    Y1 = df['Category ID'][GNG_slice]
    Y2 = df['Category ID'][LR_slice]
    
    M1 = np.stack(df['Spike rate'][GNG_slice])
    #X1 = standardize(M1)
    X1 = standardize(M1@pca_m)
    
    M2 = np.stack(df['Spike rate'][LR_slice])
    #X2 = standardize(M2)
    X2 = standardize(M2@pca_m)

    pipe = Pipeline([('scaler',  StandardScaler()),('svc', SVC())])
    rkf = RepeatedKFold(n_splits = 2, n_repeats = 10)
    fit1 = cross_validate(pipe,X1,Y1,cv=rkf,error_score=0, return_estimator=True)
    fit2 = cross_validate(pipe,X2,Y2,cv=rkf,error_score=0, return_estimator=True)

    return [fit1['test_score'],fit2['test_score']]



scores_cat = []
for f in range(len(files)) :
    print(f)
    d = data_full[files[f]]
    pca_mat = pca_s[f].T
    scores_cat.append(decoding_cat(d,pca_mat))

scores_cat = np.array(scores_cat)
```

## Choice decoding

```python
from sklearn.model_selection import KFold, GridSearchCV, RepeatedKFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC



def decoding_choice(df,pca_m) :
    GNG_slice = (df['Context'] == 0)
    LR_slice = (df['Context'] == 1)&(df['Choice'] != 0)

    Y1 = df['Choice'][GNG_slice]
    Y2 = df['Choice'][LR_slice]
    
    M1 = np.stack(df['Spike rate'][GNG_slice])
    #X1 = standardize(M1)
    X1 = standardize(M1@pca_m)
    
    M2 = np.stack(df['Spike rate'][LR_slice])
    #X2 = standardize(M2)
    X2 = standardize(M2@pca_m)

    pipe = Pipeline([('scaler',  StandardScaler()),('svc', SVC())])
    rkf = RepeatedKFold(n_splits = 2, n_repeats = 10)
    fit1 = cross_validate(pipe,X1,Y1,cv=rkf,error_score=0, return_estimator=True)
    fit2 = cross_validate(pipe,X2,Y2,cv=rkf,error_score=0, return_estimator=True)

    return [fit1['test_score'],fit2['test_score']]



scores_choice = []
for f in range(len(files)) :
    print(f)
    d = data_full[files[f]]
    pca_mat = pca_s[f].T
    scores_choice.append(decoding_choice(d,pca_mat))

scores_choice = np.array(scores_choice)
```

## Task decoding

```python
from sklearn.model_selection import KFold, GridSearchCV, RepeatedKFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC



def decoding_task(df,pca_m) :
    Y = df['Context']
    
    M = np.stack(df['Spike rate'])
    #X1 = standardize(M1)
    X = standardize(M@pca_m)

    pipe = Pipeline([('scaler',  StandardScaler()),('svc', SVC())])
    rkf = RepeatedKFold(n_splits = 2, n_repeats = 10)
    fit = cross_validate(pipe,X,Y,cv=rkf,error_score=0, return_estimator=True)

    return fit['test_score']



scores_task = []
for f in range(len(files)) :
    print(f)
    d = data_full[files[f]]
    pca_mat = pca_s[f].T
    scores_task.append(decoding_task(d,pca_mat))

scores_task = np.array(scores_task)
```

```python
def threeD_plot(d,shape=(1,1),dot_colors=['green','grey','black','lime','blue','red','cyan','lightpink'], \
                light=False,gif=False,gif_name='pseudopop_3D_view.gif',hyperplans = [],curves = []) :
    
    """ Generates 3D plots to visualize data points in some space for each label and animal
    params: data_files : Files/Animals to plot
    shape : Need to match the number of animals (i.e len(data_files))
    dot_colors : Format : 'Go/Hit, CR, FA, Miss, Right, Left, NoGo/Right, NoGo/Left'
    hyperplans : List of hyperplans to plot. Must be of shape (len(data_files) by 2) first element correspond to the hyperplan offset. Second element to the hyperplan normal vector.
    curves : List of curves to plot. Must be of shape(len(data_files) by number of curves by number of components in the space)
    """
    pv.set_jupyter_backend('client')
    p = pv.Plotter(shape=shape, window_size = (600,400))
    p.set_background('white')
    
    GNG_trials = d['Context'] == 0
    LR_trials = d['Context'] == 1
    t_idx = produce_trial_index(d)
    

    go_hit_idx = t_idx['Go / Go : Hit']
    nogo_error_idx = t_idx['Go / Nogo : Miss']

    nogo_hit_idx = t_idx['NoGo / NoGo : CR']
    go_error_idx =  t_idx['NoGo / Go : FA']


    left_hit_idx =  t_idx['Left / Left']
    snooze_left_idx = t_idx['Left / Snooze']
    left_error_idx = t_idx['Left / Right']

    right_hit_idx = t_idx['Right / Right']
    right_error_idx = t_idx['Right / Left']
    snooze_right_idx =  t_idx['Right / Snooze']
    
    D = np.stack(d['Spike rate'])
    D = standardize(D)
    M = pca1.transform(D)
    
    
    pcs_numbers = [0,2,3]
    
    M = M[:,pcs_numbers]
    
    #M ,pca = PCA_decomposition(d, return_model=True)

    p.subplot(0,0)
    
    p.set_position(np.mean(M[GNG_trials,:3],axis=0))

    #----- This block generates the sphere for each trial label and render them in the 3D plot----
    radius = 0.2
    alpha = 1
    
    go_hit_spheres = [pv.Sphere(center = M[go_hit_idx][i,:3], radius = radius) for i in range(np.shape(M[go_hit_idx])[0])]
    nogo_hit_spheres = [pv.Sphere(center = M[nogo_hit_idx][i,:3], radius = radius) for i in range(np.shape(M[nogo_hit_idx])[0])]

    go_error_spheres = [pv.Sphere(center = M[go_error_idx][i,:3], radius = radius) for i in range(np.shape(M[go_error_idx])[0])]
    nogo_error_spheres = [pv.Sphere(center = M[nogo_error_idx][i,:3], radius = radius) for i in range(np.shape(M[nogo_error_idx])[0])]

    right_spheres = [pv.Sphere(center = M[right_hit_idx][i,:3], radius = radius) for i in range(np.shape(M[right_hit_idx])[0])]
    left_spheres = [pv.Sphere(center = M[left_hit_idx][i,:3], radius = radius) for i in range(np.shape(M[left_hit_idx])[0])]

    right_error_spheres = [pv.Sphere(center = M[right_error_idx][i,:3], radius = radius) for i in range(np.shape(M[right_error_idx])[0])]
    left_error_spheres = [pv.Sphere(center = M[left_error_idx][i,:3], radius = radius) for i in range(np.shape(M[left_error_idx])[0])]

    snooze_right_spheres = [pv.Sphere(center = M[snooze_right_idx][i,:3], radius = radius) for i in range(np.shape(M[snooze_right_idx])[0])]
    snooze_left_spheres = [pv.Sphere(center = M[snooze_left_idx][i,:3], radius = radius) for i in range(np.shape(M[snooze_left_idx])[0])]
    
    for go_hit_sphere in go_hit_spheres :
        p.add_mesh(go_hit_sphere, color=dot_colors[0], opacity = alpha, lighting=light)    

    for nogo_hit_sphere in nogo_hit_spheres :
        p.add_mesh(nogo_hit_sphere, color=dot_colors[1], opacity = alpha, lighting=light)

    for go_error_sphere in go_error_spheres :
        p.add_mesh(go_error_sphere, color=dot_colors[2], opacity = alpha, lighting=light) 

    for nogo_error_sphere in nogo_error_spheres :
        p.add_mesh(nogo_error_sphere, color=dot_colors[3], opacity = alpha, lighting=light)

    for right_sphere in right_spheres :
        p.add_mesh(right_sphere, color=dot_colors[4], opacity = alpha, lighting=light)

    for left_sphere in left_spheres :
        p.add_mesh(left_sphere, color=dot_colors[5], opacity = alpha, lighting=light)
    
    for snooze_right_sphere in snooze_right_spheres :
        p.add_mesh(snooze_right_sphere, color=dot_colors[6], opacity = alpha, lighting=light)

    for snooze_left_sphere in snooze_left_spheres :
        p.add_mesh(snooze_left_sphere, color=dot_colors[7], opacity = alpha, lighting=light)

    """
    for right_error_sphere in right_error_spheres :
        p.add_mesh(right_error_sphere, color=dot_colors[8],  opacity = alpha, lighting=light)

    for left_error_sphere in left_error_spheres :
        p.add_mesh(left_error_sphere, color=dot_colors[9],  opacity = alpha, lighting=light)
    """
    #-----END------

    #Plot the hyperplans if they have been given in the function call
    if len(hyperplans) > 0 :
        print('OK')
        hyperplan = hyperplans[c]
        p.add_mesh(pv.Plane(center=hyperplan[0],direction=hyperplan[1:],i_size=15,j_size=15),color='brown',opacity=0.1)



    #Plot the curves if they have been given in the function call
    if np.shape(curves)[0] > 0 :
        print('OK')
        curve_set = curves[:,:]
        a = 0
        colors = ['orange','gold','orange','gold']
        for curve in curve_set :
            print(a)
            if a < 2 :
                offset = np.mean(M[GNG_trials,:3],axis=0)
            if a >= 2 :
                offset = np.mean(M[LR_trials,:3],axis=0)
            p.add_mesh(pv.Line(offset - 6*curve[pcs_numbers]/np.linalg.norm(curve[pcs_numbers]),offset + 6*curve[pcs_numbers]/np.linalg.norm(curve[pcs_numbers])),color=colors[a], line_width=3)
            a+=1


    p.set_position(np.array([20,10,40]))
    p.camera.zoom(0.2)
    #p.set_position(np.array([10,30,-15]))
    p.show()

        
```
